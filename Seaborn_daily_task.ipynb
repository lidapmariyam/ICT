```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Violinplot

```python
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title('species vs sepal length')
plt.show()
```

Swarmplot

```python
sns.swarmplot(x="species", y="sepal_length", data=iris)
plt.title('Species vs sepal length')
plt.show()
```

Stripplot

```python
sns.stripplot(x="species", y="sepal_width", data=iris)
plt.title('Species vs sepal width')
plt.show()
```

Boxplot

```python
sns.boxplot(x="species", y="petal_length", data=iris)
plt.title('Species vs petal length')
plt.show()
```

Barplot

```python
sns.barplot(x="species", y="petal_width", data=iris)
plt.title('Species vs petal width')
plt.show()
```

Countplot

```python
sns.countplot(x='species', data=iris)
plt.show()
```

Heatmap

```python
iris_numeric = iris.drop(columns='species')
corr = iris_numeric.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset')
plt.show()
```

Scatterplot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='Set1')
plt.title('Scatter Plot of Sepal Length vs Sepal Width by Species')
plt.legend(title='Species')
plt.show()
```

Line plot

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='species', y='sepal_length', data=iris, marker='o', hue='species')
plt.title('Line Plot of Sepal Length for Each Iris Species')
plt.legend(title='Species')
plt.show()
```
