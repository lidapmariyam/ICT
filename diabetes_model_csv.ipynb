```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/diabetes_model.csv')
```

```python
df.head()
```

```python
df.describe()
```

univariate analysis
1)histogram
boxplot

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Glucose'], bins=30, kde=True)
plt.title('Histogram of Glucose Levels')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['BloodPressure'], bins=30, kde=True)
plt.title('Histogram of BloodPressure Levels')
plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Insulin'], bins=30, kde=True)
plt.title('Histogram of Insulin Levels')
plt.xlabel('Insulin')
plt.ylabel('Frequency')
plt.show()
```

2) box plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['BMI'])
plt.title('Box Plot of BMI')
plt.xlabel('BMI')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age'])
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Glucose'])
plt.title('Box Plot of Glucose')
plt.xlabel('Glucose')
plt.show()
```

Bivariate analysis
1)scatter plot


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Glucose'], y=df['BMI'])
plt.title('Scatter Plot of Glucose vs BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Insulin'], y=df['BMI'])
plt.title('Scatter Plot of Insulin vs BMI')
plt.xlabel('Insulin')
plt.ylabel('BMI')
plt.show()
```

2)correlation matrices

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

Multivariate analysis
1)pair plot


```python
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'Insulin', 'Outcome']], hue='Outcome')
plt.show()
```

2) 3D plot

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Glucose'], df['BMI'], df['Age'], c=df['Outcome'], cmap='coolwarm')
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
plt.title('3D Scatter Plot of Glucose, BMI, and Age')
plt.show()
```
